# Cryptocurrency Price Prediction

You must have heard or invested in any cryptocurrency once in life. It is global medium of exchange that is encrypted and decentralized. Many people used cyptocurrencies as a form of investing because it gives returns even in a short period. Bitcoin, Ethereum and Binance Coin are among the popular cryptocurrencies today.

Predicting the price of cryptocurrencies is one of the popular case studies in the data science community. The price of stocks and cryptocurrencies don't just depend on the number of people who buy or sell them. Today, the change in the price of the investments also depends on the changes in the financial policies of the government regarding any cryptocurrencies. The feelings of people towards a particular cryptocurrency or personality who directly or indirectly endorse a cryptocurrency also result in a huge buying and selling of a particular crytocurrency, resulting in a change in prices.

In short, buying and selling result in a change in the price of any cryptocurrency, but having and selling trends on many factors. Using machine learning for cryptocurrency price prediction can also work in situations where price change due to historical prices that people see before buying and selling their cryptocurrency.

```python
import pandas as pd
import numpy as np
from datetime import date, timedelta
import yfinance as yf
import plotly.graph_objects as go
from autots import AutoTS
```

```python
today= date.today()

d1= today.strftime("%Y-%m-%d")
end_date= d1
d2= date.today() - timedelta(days=730)
d2= d2.strftime("%Y-%m-%d")
start_date= d2
```

```python
data= yf.download("BTC-USD", start= start_date, end= end_date, progress=False)
```

```python
data["Date"]= data.index
data= data[["Date", "Open", "High", "Low", "Close", "Adj Close", "Volume"]]
data.reset_index(drop=True, inplace=True)
data.head()
```

In the above code, collected the latest data of Bitcoin prices for the past 730 days, and then i have prepared it for any data science task. Now, let's have a look at the shape of this dataset to see if we are looking with 730 rows or not:

```python
data.shape
```

So the dataset contains 730 rows, where the first row contains the name of each column. Now let's visualize the change in bitcoin prices till today by using a candlestick chart:

```python
figure= go.Figure(data= [go.Candlestick(x= data["Date"], open= data["Open"], high= data["High"],
                                        low= data["Low"], close= data["Close"])])

figure.update_layout(title= "Bitcoin Price Analysis")
figure.show()
```

The close column in the dataset contains the values we need to predict. So let's have a look at the correlation of all the columns in the data concerning the close column:

```python
Correlation = data.corr()
print(Correlation["Close"].sort_values(ascending=False))
```

Predicting the future prices of cryptocurrency is based on the problem of the Time series analysis. The AutoTS library in python is one of the best libraries for time series analysis. So here using the AutoTS library to predict the bitcoin prices for the next 30 days:

```python
model= AutoTS(forecast_length= 30, frequency= "infer", ensemble= "simple", max_generations=5)
model= model.fit(data, date_col= "Date", value_col= "Close", id_col= None)
prediction= model.predict()
forecast= prediction.forecast
```
