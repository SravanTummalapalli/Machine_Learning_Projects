# Covid-19 Death Prediction

Covid-19 is one of the deadlist virus you have ever heard. Mutations in covid-19 make it either more deadly or more infectious. We have seen a lot of deaths from covid-19 while there is a higher wave of cases. We can use historical data on covid-19 cases and deaths to predict the number of deaths in the future.

You are given a dataset of covid-19 in india from 30 January 2020 to 18 January 2022. The datset contains information about the daily confirmed cases and deaths. Below are all the columns of the dataset:

1. <b>Date:</b> Contain the date of the record
2. <b>Date_YMD:</b> Contains date in Year-Month Format
3. <b>Daily Confirmed</b> Contains the daily confirmed cases of Covid-19
4. <b>Daily Deceased</b> Contains the daily deaths due to Covid-19

You need to use this historical data of Covid-19 cases and deaths to predict the number of deaths for the next 30 days

```python
import pandas as pd
import numpy as np
import plotly.express as px
from autots import AutoTS
```

```python
# URL of the CSV file
url = "https://raw.githubusercontent.com/amankharwal/Website-data/master/COVID19%20data%20for%20overall%20INDIA.csv"

# Read the CSV file from the URL
data = pd.read_csv(url)

# Save the file locally
data.to_csv("covid19death.csv", index=False)
```

```python
data=pd.read_csv("covid19death.csv")
```

```python
data.head()
```

Before moving forward, let's have a quick look at whether this dataset contains any null values or not:

```python
data.isnull().sum()
```

We don't need the date column, so let's drop this column from our dataset:

```python
data= data.drop("Date", axis=1)
```

Let's have a look at the daily confirmed cases of covid-19:

```python
fig=px.bar(data, x="Date_YMD", y="Daily Confirmed")
fig.show()
```

In the data visulaization above, we can see a high wave of covid-19 cases between April 2021 and May 2021.


Now let's visualize the death rate due to covid-19:

```python
cases= data["Daily Confirmed"].sum()
deceased= data["Daily Deceased"].sum()

labels= ["Confirmed", "Deceased"]
values= [cases, deceased]

fig= px.pie(data, values=values, names= labels, title="Daily Cases Vs Daily Deaths", hole=0.5)

fig.show()
```

Now let's caluclate the death rate of Covid-19:

```python
death_rate= (data["Daily Deceased"].sum()/ data["Daily Confirmed"].sum())*100
print(death_rate)
```

Now let's have a look at the daily deaths of Covid-19:

```python
fig= px.bar(data, x="Date_YMD", y="Daily Deceased")
fig.show()
```

We can see a high number of deaths during the high wave of covid-19 cases.

Now let's move to the task of covid-19 deaths prediction for the next 30 days. Here i will be using the AutoTS library, which is one of the best Automatic Machine Learning libraries for Time Series Analysis.

```python
model= AutoTS(forecast_length=30, frequency="infer", ensemble="simple")
model=model.fit(data, date_col="Date_YMD", value_col="Daily Deceased", id_col=None)
prediction= model.predict()
```

```python
forecast= prediction.forecast
print(forecast)
```
